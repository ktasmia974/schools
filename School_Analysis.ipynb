## 📊 Step 4 — Data Visualizations

```python
def add_stamp(ax):
    """Add student info watermark to every chart."""
    ax.text(0.99, 0.01, 'Tasmia Khan | SAP: 80412 | DAA | Sir Karar Haider',
            transform=ax.transAxes, fontsize=7.5, color='#555555',
            ha='right', va='bottom', alpha=0.75,
            bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.5))

print('✅ add_stamp() helper defined — every chart will carry your name.')
```

```python
# ── Chart 1 : Annual Fee Comparison ──────────────────────────────
fig, ax = plt.subplots(figsize=(14, 8))
colors = ['#e63946' if f >= 60000 else '#1a1a2e' for f in df['Annual_Fee_USD']]
bars = ax.barh(df['School_Name'], df['Annual_Fee_USD'], color=colors, edgecolor='white', linewidth=0.5)
ax.set_xlabel('Annual Fee (USD)', fontsize=12)
ax.set_title('🏫  Schools — Annual Fee Comparison\nTasmia Khan | SAP: 80412 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold', pad=12)
ax.axvline(60000, color='#e63946', linestyle='--', alpha=0.5, label='$60,000 Mark')
for bar, val in zip(bars, df['Annual_Fee_USD']):
    ax.text(val + 1000, bar.get_y() + bar.get_height() / 2,
            f'${val:,}', va='center', fontsize=7.5)
ax.legend(); add_stamp(ax); plt.tight_layout()
plt.savefig('chart1_fee_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 1 saved.')
```

```python
# ── Chart 2 : Fee vs Global Ranking ──────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))
sc = ax.scatter(df['Global_Ranking'], df['Annual_Fee_USD'] / 1000,
                c=df['Founded'], cmap='plasma', s=110, edgecolors='white', linewidth=0.8)
for _, row in df.iterrows():
    ax.annotate(row['School_Name'], (row['Global_Ranking'], row['Annual_Fee_USD'] / 1000),
                textcoords='offset points', xytext=(6, 4), fontsize=7, alpha=0.85)
plt.colorbar(sc, ax=ax, label='Founded Year')
ax.set_xlabel('Global Ranking (1 = best)', fontsize=12)
ax.set_ylabel('Annual Fee (Thousand USD)', fontsize=12)
ax.invert_xaxis()
ax.set_title('💰  Fee vs Global Ranking\nTasmia Khan | SAP: 80412 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
add_stamp(ax); plt.tight_layout()
plt.savefig('chart2_fee_vs_ranking.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 2 saved.')
```

```python
# ── Chart 3 : Top 10 Best-Ranked Schools ─────────────────────────
top10 = df.nsmallest(10, 'Global_Ranking')
fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(top10['School_Name'], top10['Global_Ranking'],
              color=sns.color_palette('rocket_r', len(top10)), edgecolor='white', linewidth=0.6)
ax.set_ylabel('Global Ranking (lower = better)', fontsize=12)
ax.set_title('🏆  Top 10 Best-Ranked Schools\nTasmia Khan | SAP: 80412 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
plt.xticks(rotation=38, ha='right', fontsize=9)
for bar, val in zip(bars, top10['Global_Ranking']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.2,
            f'#{val}', ha='center', fontsize=8)
add_stamp(ax); plt.tight_layout()
plt.savefig('chart3_top_ranked.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 3 saved.')
```

```python
# ── Chart 4 : Student-Teacher Ratio ──────────────────────────────
ratio_sorted = df.sort_values('Student_Teacher_Ratio')
fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.barh(ratio_sorted['School_Name'], ratio_sorted['Student_Teacher_Ratio'],
               color=sns.color_palette('crest', len(ratio_sorted)), edgecolor='white', linewidth=0.5)
ax.set_xlabel('Student-Teacher Ratio', fontsize=12)
ax.set_title('👩‍🏫  Student-Teacher Ratio Comparison\nTasmia Khan | SAP: 80412 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
for bar, val in zip(bars, ratio_sorted['Student_Teacher_Ratio']):
    ax.text(val + 0.2, bar.get_y() + bar.get_height() / 2, f'{val}', va='center', fontsize=8)
add_stamp(ax); plt.tight_layout()
plt.savefig('chart4_student_teacher_ratio.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 4 saved.')
```

```python
# ── Chart 5 : Correlation Heatmap ────────────────────────────────
num = df[['Students','Teachers','Annual_Fee_USD','Campus_Acres','Global_Ranking','Extracurricular_Activities','Rating_Out_Of_5']]
corr = num.corr()
fig, ax = plt.subplots(figsize=(10, 7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, ax=ax, vmin=-1, vmax=1)
ax.set_title('🔗  Correlation Matrix — Scale, Fee & Ranking\nTasmia Khan | SAP: 80412 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=11, fontweight='bold')
add_stamp(ax); plt.tight_layout()
plt.savefig('chart5_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 5 saved.')
```
